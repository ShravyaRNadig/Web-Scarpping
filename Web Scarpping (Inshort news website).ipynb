```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
data=requests.get('https://inshorts.com/en/read')
data
```

```python
d=BeautifulSoup(data.content,'html.parser')
##csv parser
## since its html parser
```

```python
d
```

```python
headline=d.find_all('span',attrs={'itemprop':'headline'})
headline
##headline.text error cz its list
```

```python
docs=[]
for news in headline:
    docs.append(news.text)
docs
```

```python
df=pd.DataFrame(docs,columns=["Headlines"])
```

```python
df
```

```python
article=d.find_all('div',attrs={'itemprop':'articleBody'})
article
```

```python
art=[]
for news in article:
    art.append(news.text)
art
```

```python
df['Article']=art
df
```
