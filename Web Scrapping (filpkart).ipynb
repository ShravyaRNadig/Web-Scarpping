## Fetching of data


```python
#most importent in fetching the data is url
url="https://www.flipkart.com/search?q=mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&as-pos=1&as-type=RECENT&suggestionId=mobiles%7CMobiles&requestId=894d4979-f05c-47b5-85ea-45c8bc50d600&as-searchtext=mobile"
data=requests.get(url)
data
```

## Parsing of Data

```python
soup=BeautifulSoup(data.content,"html.parser")
## to extract html content
```

```python
product=[]
price=[]
ratings=[]
for a in soup.find_all('a',attrs={'class':'_1fQZEK'}):
    name=a.find('div',attrs={'class':'_4rR01T'})
    product.append(name.text)
product
```

```python
product=[]
price=[]
ratings=[]
for a in soup.find_all('a',attrs={'class':'_1fQZEK'}):
    name=a.find('div',attrs={'class':'_4rR01T'})
    p=a.find('div',attrs={'class':'_3tbKJL'})
    product.append(name.text)
    price.append(p.text)

price
```

```python
product
```

### Stroing The Data

```python
df=pd.DataFrame({'Product':product,'Price':price})
df
```
